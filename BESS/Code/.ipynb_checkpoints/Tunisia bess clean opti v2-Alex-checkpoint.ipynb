```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns # statistical data visualization
import math #for number e
import numpy_financial as npf #for NPV
import plotly.graph_objects as go

##Cases and scenarios

Source='Stand_alone' #'Solar' #'Stand_alone' #'Wind'
scenario=1
marginal_technology='base_curtail' #base_curtail #doble #hydrogen #interconnection
capacity=100 #[MW] 
time=4 #[hours]

## NPV
#Sensitivity variables
inv_c1=287000#0.90 #investment constant MWh [USD/MWh]
inv_c2=288000#0.90 #investment constant MW  [USD/(MW)]
OM=36000#*0.9 O&M yearly value [USD/(MW*year)]
inflation = 0.05#0.1 #0.05 [%]
discount_rate = 0.08#0.08 #0.04 #https://tradingeconomics.com/tunisia/interest-rate
##
price_increase = 0.0 #tariff or price increase per year, if increases are not annual this must be modelled
life_project = 10 #Financial analysis period, see replacement [years]
##

inputs = pd.ExcelFile('/srv/share/__Mitarbeiter/nescobosa/compilation.xlsx')
gen_cost_input=pd.ExcelFile('/srv/share/__Mitarbeiter/nescobosa/Tunisia_results/Gen_cost/'+marginal_technology+'.xlsx')

generation_inputs=inputs.parse('Generation')
gen_cost=gen_cost_input.parse('Sheet1',index_col=0)

if scenario == 1:
    size_proportion=1
elif scenario == 2:
    size_proportion= 0.7
elif scenario==3:
    size_proportion=0.5
elif scenario==4:
    size_proportion=0.3


if Source == 'Stand_alone':
    generation_inputs[Source]=[0]*8760
else:
    generation_inputs[Source]=capacity*generation_inputs[Source+'_CF'] 

eta=0.927 #charging/discharging efficiency [%]
maxDod=0.90 #Maximum depth of discharge [%]
EOL=0.80 # total degradation end of life [%]
life_BESS=6000 #number of cycles [cycles]
sp_deg=(1-EOL)/life_BESS #Specific degradation per kWh used
days=365 #operational days 


##

#Sizing
C_P=round(capacity*size_proportion/eta) #Determining Power BESS capacity considering efficiency
C_En=round(C_P*time/maxDod) #Determining Energy BESS capacity needed for the future considering max DoD (only need to consider discharging losses since charging losses are not stored in BESS)
C_E=C_En #C_E is used as the BESS capacity that is being degraded and C_En is the original capacity
BESS_results={}
num_bat=[0]
```

```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory, SolverStatus, TerminationCondition
import copy

def DispatchBattery (time_duration,generation, price, socMax, soc0,DoD,battery_power,efficiency):
    timeSteps = range (time_duration)
    assert len(timeSteps)==len(generation)
    assert len(timeSteps)==len(price)
    model = pyo.ConcreteModel()
    __define_Variables__ (model,timeSteps, socMax, battery_power,DoD)
    __define_Constraints__ (model,generation,soc0, socMax,efficiency)
    model.profit = pyo.Objective(expr=sum ([model.powerToGrid [t] * price [t] for t in timeSteps]), sense=pyo.maximize)
    
    solver = SolverFactory('glpk')
    results = solver.solve(model)
    if (results.solver.status != SolverStatus.ok) or (results.solver.termination_condition != TerminationCondition.optimal):
        return False
    
    batteryPower=[model.chargePower[t].value-model.dischargePower[t].value for t in timeSteps]
    powerToGrid = [model.powerToGrid [t].value for t in timeSteps]
    soc = [model.soc [t].value for t in timeSteps]
    profit = copy.deepcopy (pyo.value(model.profit))

    return batteryPower, powerToGrid, soc, profit

def __define_Variables__ (model, timeSteps, socMax,battery_power,DoD):
    if Source== 'Stand_alone':
        model.powerToGrid = pyo.Var (timeSteps, within=pyo.Reals) #stand-alone
    else:
        model.powerToGrid = pyo.Var (timeSteps, within=pyo.NonNegativeReals) #hybrid power plant
    model.chargePower = pyo.Var (timeSteps, bounds = (0,battery_power))
    model.dischargePower = pyo.Var (timeSteps, bounds = (0, battery_power))
    model.soc = pyo.Var(timeSteps, within=pyo.NonNegativeReals, bounds=(DoD, socMax))

def __define_Constraints__ (model, Generation,soc0,socMax,efficiency):
    model.src = pyo.ConstraintList()
    for t in range(len(Generation)):
        model.src.add (model.powerToGrid [t] == Generation [t] + model.dischargePower [t]*efficiency-model.chargePower[t])
        if t > 0:
            model.src.add (model.soc [t] == model.soc [t-1]- model.dischargePower [t] + model.chargePower [t]*efficiency)
        else:
            model.src.add (model.soc [t] == soc0- model.dischargePower [t] + model.chargePower [t]*efficiency)
        model.src.add (model.soc[t]-model.dischargePower[t]+efficiency*model.chargePower[t]<=socMax)    
```

```python
#Algorithm blocks load following hybrid
def algorithm(C_E,gen_cost):
    BESS=pd.DataFrame()    
    BESS['Gen']=generation_inputs[Source] #generation of the specific plant
    (charge,output,soc,profit)=DispatchBattery(8760,generation_inputs[Source],gen_cost,C_E*maxDod,(1-maxDod)*C_E,(1-maxDod)*C_E,C_P,eta)
    degrad_acc=0
    for v in range (0,len(charge)):
        degrad_acc+=abs(charge[v])/(2*C_En*maxDod)
    C_E-=C_E*degrad_acc*sp_deg
    BESS['SOC']=soc
    BESS['CH_DI']=charge
    BESS['Output']=output
    BESS['Price']=gen_cost
    BESS['BAU']=gen_cost*generation_inputs[Source]
    BESS['BESS_operation']=gen_cost*BESS['Output']
    BESS['Savings_']=gen_cost*(BESS['Output']-generation_inputs[Source])
    return(BESS,C_E,profit)


for y in range (2025,2036):
    BESS,C_E,profit=algorithm(C_E,gen_cost[y])
    print (y)
    BESS_results[y]=BESS
    if C_E<(EOL*C_En):
        num_bat.append(y-2025)
        C_E=C_En      

```

```python
def NPVf(BESS_results,investment,OM,C_E,num_bat):
    #for p in range (0,len(num_bat)):    
    annual_savings=[investment*C_P] #List to store the net annual savings considering fixed savings per day and the actual_capacity- O&M costs
    for x in range(0,life_project+1): #Cycle through the years of the project
        index=2025+x
        print('savings'+str(2025+x),BESS_results[index].Savings_.sum())
        annual_savings.append(round((BESS_results[index].Savings_.sum()*(1+price_increase)**(x+1)-OM*C_P)*(1+inflation)**(x+1),2))
    salvage=(C_En-C_E)/C_En #salvage value by the end of the project
    annual_savings.append(-salvage*investment*C_P)
    NPV=round(npf.npv(discount_rate,annual_savings),3)
    IRR=round(npf.irr(annual_savings),4)*100
    PBP=round(abs(investment*C_P)/annual_savings[1],2)
    return (NPV)

investment=-(inv_c1*time+inv_c2)
NPV=NPVf(BESS_results,investment,OM,C_E,num_bat) 
```

```python
savs=[]
x=[]
for w in range (2025,2036):
    x.append(w+1)
    savs.append(round(BESS_results[w]['Savings_'].sum(),0))

```

```python
def results(BESS,year,st,n):
    fi=st+24*n+1
    x_values=range(st,fi)
    figure,axis=plt.subplots(3,1, figsize=(10, 10))
    
    axis[0].step(x_values,BESS['Price'][st:fi],color='crimson', label='Price [USD/MW]')
    axis[0].set_ylabel('Price [USD/MW]', fontsize=14)
    axis[0].set_xticks([])
    
    axis[1].step(x_values,BESS['Output'][st:fi], color='royalblue',label='Output at interconnection point [MW]')
    axis[1].step(x_values,BESS['CH_DI'][st:fi], color='k', label='Charging/discharging of BESS [MW]')
    if Source!='Stand_alone':
        axis[1].step(x_values,BESS['Gen'][st:fi],color='gold', label='Generation [MW]')
    axis[1].set_ylabel('Power [MW]', fontsize=14)
    axis[1].set_xticks([])
    
    axis[2].step(x_values,BESS['SOC'][st:fi],color='forestgreen',label='SOC [MWh]')
    axis[2].set_ylabel('Energy [MWh]', fontsize=14)
    axis[2].set_xlabel('Hour', fontsize=14)
    
    plt.subplots_adjust(hspace=0.01)
    axis[0].legend(bbox_to_anchor=(0, 1.02,1,0.2), loc="lower left",mode="expand", borderaxespad=0,ncol=1)
    axis[1].legend(bbox_to_anchor=(0, 1.02,1,0.2), loc="lower left",mode="expand", borderaxespad=0,ncol=3)
    axis[2].legend(bbox_to_anchor=(0, 1.02,1,0.2), loc="lower left",mode="expand", borderaxespad=0,ncol=1)

    for a in axis.flatten():
        a.tick_params(axis='both', which='major', labelsize=14)
        a.tick_params(axis='both', which='minor', labelsize=12)   
    plt.tight_layout()
    plt.show()

results(BESS_results[2025],2025,3600,2)
results(BESS_results[2030],2030,3600,2)
results(BESS_results[2035],2035,3600,2)
```
